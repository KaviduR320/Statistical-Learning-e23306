#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,

Assume the probabilities are

P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad

Thus:

* $Y=0$ means “low-risk customer,”

### **1. Probability Space $(\Omega, \mathcal{F}, P)$**

* **Sample Space ($\Omega$):** $\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}$
* **$\sigma$-algebra ($\mathcal{F}$):** $\mathcal{F} = \mathcal{P}(\Omega)$ is the power set of $\Omega$. It contains all $2^4 = 16$ possible subsets of $\Omega$.
* **Probability Measure ($P$):** Defined on $\mathcal{F}$ such that for any subset $A \subseteq \Omega$, the probability $P(A) = \sum_{\omega_i \in A} P(\{\omega_i\})$. The foundational probabilities are:
* $P(\{\omega_1\}) = 0.50$
* $P(\{\omega_2\}) = 0.10$
* $P(\{\omega_3\}) = 0.20$
* $P(\{\omega_4\}) = 0.20$



### **2. The $\sigma$-algebra generated by $Y$, $\sigma(Y)$**

The variable $Y$ groups the sample space into two pre-images based on the customer's risk class:

* $Y=0$ (low-risk): $A_0 = \{\omega_1, \omega_2\}$
* $Y=1$ (high-risk): $A_1 = \{\omega_3, \omega_4\}$

The $\sigma$-algebra generated by $Y$ is the smallest $\sigma$-algebra containing these sets:


$$\sigma(Y) = \{\emptyset, \Omega, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}\}$$

### **3. Why $E[X \mid Y]$ must be $\sigma(Y)$-measurable**

By definition, the conditional expectation of $X$ given $Y$ is a random variable that relies only on the information provided by $Y$. Mathematically, this requires $E[X \mid Y]$ to be measurable with respect to $\sigma(Y)$. Because a $\sigma(Y)$-measurable function must be constant on the fundamental "atoms" (or partitions) of $\sigma(Y)$, $E[X \mid Y]$ can only take one constant value for the set $\{\omega_1, \omega_2\}$ and another constant value for the set $\{\omega_3, \omega_4\}$.

### **4. Computing $E[X \mid Y]$ explicitly**

We calculate the expected value of fraud ($X$) for each risk class ($Y$):

* **For $Y=0$:**

$$P(Y=0) = P(\{\omega_1\}) + P(\{\omega_2\}) = 0.50 + 0.10 = 0.60$$


$$E[X \mid Y=0] = \frac{X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\})}{P(Y=0)} = \frac{0(0.50) + 1(0.10)}{0.60} = \frac{1}{6}$$


* **For $Y=1$:**

$$P(Y=1) = P(\{\omega_3\}) + P(\{\omega_4\}) = 0.20 + 0.20 = 0.40$$


$$E[X \mid Y=1] = \frac{X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\})}{P(Y=1)} = \frac{0(0.20) + 1(0.20)}{0.40} = \frac{1}{2}$$



Combining these, we get the indicator function notation:


$$E[X \mid Y] = \frac{1}{6} \mathbf{1}_{\{Y=0\}} + \frac{1}{2} \mathbf{1}_{\{Y=1\}}$$

### **5. Verifying the defining property of conditional expectation**

We must show $\int_A E[X \mid Y] dP = \int_A X dP$ for all $A \in \sigma(Y)$.

* **For $A = \{\omega_1, \omega_2\}$ (where $Y=0$):**
* LHS: $\int_A E[X \mid Y] dP = \frac{1}{6} \times P(A) = \frac{1}{6} \times 0.60 = 0.10$
* RHS: $\int_A X dP = X(\omega_1)P(\omega_1) + X(\omega_2)P(\omega_2) = 0 + 1(0.10) = 0.10$


* **For $A = \{\omega_3, \omega_4\}$ (where $Y=1$):**
* LHS: $\int_A E[X \mid Y] dP = \frac{1}{2} \times P(A) = \frac{1}{2} \times 0.40 = 0.20$
* RHS: $\int_A X dP = X(\omega_3)P(\omega_3) + X(\omega_4)P(\omega_4) = 0 + 1(0.20) = 0.20$


* **For $A = \Omega$** and **$A = \emptyset$**, the integrals also perfectly match based on the additions of the above parts (e.g., $0.10 + 0.20 = 0.30$).

### **6. Finding the Conditional PMF $p_{X|Y}(x \mid y)$**

* **Given $Y = 0$:**
* $p_{X|Y}(0 \mid 0) = \frac{P(\omega_1)}{P(Y=0)} = \frac{0.50}{0.60} = \frac{5}{6}$
* $p_{X|Y}(1 \mid 0) = \frac{P(\omega_2)}{P(Y=0)} = \frac{0.10}{0.60} = \frac{1}{6}$


* **Given $Y = 1$:**
* $p_{X|Y}(0 \mid 1) = \frac{P(\omega_3)}{P(Y=1)} = \frac{0.20}{0.40} = \frac{1}{2}$
* $p_{X|Y}(1 \mid 1) = \frac{P(\omega_4)}{P(Y=1)} = \frac{0.20}{0.40} = \frac{1}{2}$



### **7. Verifying the Identity**

We check if $E[X \mid Y=y] = \sum_{x \in \{0, 1\}} x \cdot p_{X|Y}(x \mid y)$:

* **For $y=0$:** $\sum = 0 \cdot \frac{5}{6} + 1 \cdot \frac{1}{6} = \frac{1}{6}$. This matches $E[X \mid Y=0]$.
* **For $y=1$:** $\sum = 0 \cdot \frac{1}{2} + 1 \cdot \frac{1}{2} = \frac{1}{2}$. This matches $E[X \mid Y=1]$.

### **8. Calculating Mean Square Error (MSE)**

The MSE of the conditional expectation estimator is $E[Var(X \mid Y)]$.
Since $X$ is binary, $Var(X|Y) = p(1-p)$ where $p = E[X|Y]$.

* $Var(X \mid Y=0) = \frac{1}{6} \times (1 - \frac{1}{6}) = \frac{5}{36}$
* $Var(X \mid Y=1) = \frac{1}{2} \times (1 - \frac{1}{2}) = \frac{1}{4}$

$$MSE = Var(X \mid Y=0)P(Y=0) + Var(X \mid Y=1)P(Y=1)$$

$$MSE = \frac{5}{36}(0.60) + \frac{1}{4}(0.40) = \frac{3}{36} + \frac{1}{10} = \frac{1}{12} + \frac{1}{10} = \frac{11}{60} \approx 0.1833$$

### **9. Change in MSE & Law of Total Variance**

If the bank ignores $Y$, the estimate is just $E[X] = 0.30$.

* **New MSE:** This is simply the unconditional variance of $X$, $Var(X)$.

$$Var(X) = E[X^2] - E[X]^2 = 0.30 - (0.30)^2 = 0.30 - 0.09 = 0.21 = \frac{63}{300}$$


* **Relationship (Law of Total Variance):** $Var(X) = E[Var(X \mid Y)] + Var(E[X \mid Y])$. Let's check the missing piece, the variance of the conditional means:

$$Var(E[X \mid Y]) = E[(E[X|Y] - E[X])^2]$$


$$Var(E[X \mid Y]) = (\frac{1}{6} - 0.30)^2(0.60) + (\frac{1}{2} - 0.30)^2(0.40) = \frac{8}{300}$$



We calculated $MSE_{conditional} = \frac{11}{60} = \frac{55}{300}$.
$\frac{55}{300} + \frac{8}{300} = \frac{63}{300} = 0.21$. The law completely bridges the gap between the two errors.

### **10. Entropy and Mutual Information**

The marginal probabilities are $P(X=0)=0.7$, $P(X=1)=0.3$.

* **a) Entropy $H(X)$:**

$$H(X) = -[0.7 \log_2(0.7) + 0.3 \log_2(0.3)] \approx 0.8813 \text{ bits}$$


* **b) Conditional Entropy $H(X \mid Y)$:**

$$H(X \mid Y=0) = -[\frac{5}{6} \log_2(\frac{5}{6}) + \frac{1}{6} \log_2(\frac{1}{6})] \approx 0.6500 \text{ bits}$$


$$H(X \mid Y=1) = -[\frac{1}{2} \log_2(\frac{1}{2}) + \frac{1}{2} \log_2(\frac{1}{2})] = 1.000 \text{ bits}$$


$$H(X \mid Y) = P(Y=0)H(X|Y=0) + P(Y=1)H(X|Y=1) \approx 0.6(0.65) + 0.4(1.0) = 0.7900 \text{ bits}$$


* **c) Mutual Information $I(X;Y)$:**

$$I(X;Y) = H(X) - H(X \mid Y) \approx 0.8813 - 0.7900 = 0.0913 \text{ bits}$$


* **d) Percentage of Uncertainty Removed:**

$$\frac{I(X;Y)}{H(X)} = \frac{0.0913}{0.8813} \approx 10.36\%$$



Knowing the customer's risk class resolves about **10.36%** of the total uncertainty regarding whether the transaction is fraudulent.

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$



The most elegant and statistically intuitive way to prove this is by using a linear transformation to create a new random variable that is independent of $Y$.

### **1: Construct an Independent Random Variable**

We define a new random vector $W$ such that $W$ is a linear combination of $X$ and $Y$, but is **uncorrelated** (and therefore, due to joint normality, **independent**) of $Y$.

Let's define $W$ as:


$$W = X - A Y$$


where $A$ is an $n \times m$ matrix to be determined.

To make $W$ and $Y$ uncorrelated, their covariance must be zero:


$$Cov(W, Y) = Cov(X - AY, Y) = 0$$

Using the linearity of covariance:


$$Cov(X, Y) - A \cdot Cov(Y, Y) = 0$$

$$\Sigma_{XY} - A \Sigma_{YY} = 0$$

$$A = \Sigma_{XY}\Sigma_{YY}^{-1}$$

So, our independent random variable is $W = X - \Sigma_{XY}\Sigma_{YY}^{-1}Y$.

### **2: Find the Distribution of $W$**

Since $W$ is a linear combination of jointly Gaussian variables, $W$ itself is normally distributed. Let's find its mean and covariance matrix.

**Mean of $W$:**


$$E[W] = E[X] - A E[Y]$$

$$E[W] = \mu_X - \Sigma_{XY}\Sigma_{YY}^{-1}\mu_Y$$

**Variance (Covariance Matrix) of $W$:**


$$Var(W) = Cov(X - AY, X - AY)$$


Expanding this using the distributive property of covariance:


$$Var(W) = Var(X) - A Cov(Y, X) - Cov(X, Y)A^T + A Var(Y) A^T$$

Substitute $A = \Sigma_{XY}\Sigma_{YY}^{-1}$, $Var(X) = \Sigma_{XX}$, $Var(Y) = \Sigma_{YY}$, and $Cov(X,Y) = \Sigma_{XY}$:


$$Var(W) = \Sigma_{XX} - (\Sigma_{XY}\Sigma_{YY}^{-1})\Sigma_{YX} - \Sigma_{XY}(\Sigma_{YY}^{-1}\Sigma_{YX}) + (\Sigma_{XY}\Sigma_{YY}^{-1})\Sigma_{YY}(\Sigma_{YY}^{-1}\Sigma_{YX})$$

Notice that the last term simplifies to $\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$, which cancels out the middle term:


$$Var(W) = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$$

This matches the definition given in the problem: $Var(W) = \Sigma_{X \mid Y}$.
Thus, the unconditional distribution of $W$ is $\mathcal{N}(\mu_X - \Sigma_{XY}\Sigma_{YY}^{-1}\mu_Y, \Sigma_{X|Y})$.

### **3: Relate $W$ back to the Conditional Distribution $X \mid Y$**

From our definition in Step 1, we can express $X$ as:


$$X = W + AY = W + \Sigma_{XY}\Sigma_{YY}^{-1}Y$$

We want to find the distribution of $X$ **given** that $Y = y$. Because $W$ and $Y$ are independent, conditioning on $Y=y$ does not change the distribution of $W$. It only turns $Y$ into a constant $y$.

Therefore, conditionally on $Y=y$, $X$ is simply the random variable $W$ shifted by a constant vector $Ay$:


$$X \mid (Y=y) = W + \Sigma_{XY}\Sigma_{YY}^{-1}y$$

conditional mean ($\mu^*$):


$$\mu^* = E[X \mid Y=y] = E[W] + \Sigma_{XY}\Sigma_{YY}^{-1}y$$

$$\mu^* = (\mu_X - \Sigma_{XY}\Sigma_{YY}^{-1}\mu_Y) + \Sigma_{XY}\Sigma_{YY}^{-1}y$$

$$\mu^* = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y)$$


*(This exactly matches the target $\mu^*$ in the problem).*

Conditional variance:


$$Var(X \mid Y=y) = Var(W + \text{constant})$$

$$Var(X \mid Y=y) = Var(W) = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX} = \Sigma_{X \mid Y}$$


*(This exactly matches the target $\Sigma_{X|Y}$ in the problem).*

### **4: Write the Final Conditional Density Function**

Because $X \mid Y=y$ is merely a shifted Gaussian vector, it remains multivariate normal.

Plugging our derived conditional mean ($\mu^*$) and conditional covariance matrix ($\Sigma_{X|Y}$) into the standard formulation of a multivariate normal density function over $\mathbb{R}^n$ dimensions, we get:

$$f_{X|Y}(x \mid y) = \frac{1}{(2\pi)^{n/2} |\Sigma_{X|Y}|^{1/2}} \exp\left(-\frac{1}{2}(x-\mu^*)^T \Sigma_{X|Y}^{-1} (x-\mu^*)\right)$$.

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

**5. Information Theory: Differential Entropy**
* Compute the differential entropy $h(X)$.
* Compute the conditional differential entropy $h(X \mid Y)$.
* *Note: The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.*

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?
* How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).





### **1. Joint Distribution**

We are given two independent random variables: the hidden temperature $X \sim \mathcal{N}(0, 1)$ and the mechanical load (noise) $Q \sim \mathcal{N}(0, 4)$. The sensor measures $Y = X + Q$.

* **Distribution of $Y$:**
Since $Y$ is the sum of two independent normally distributed variables, $Y$ itself is normally distributed.
* Mean: $E[Y] = E[X] + E[Q] = 0 + 0 = 0$
* Variance: $Var(Y) = Var(X + Q) = Var(X) + Var(Q) = 1 + 4 = 5$
Thus, $Y \sim \mathcal{N}(0, 5)$.


* **Covariance $Cov(X, Y)$:**

$$Cov(X, Y) = Cov(X, X + Q)$$



Using the linearity of covariance:

$$Cov(X, Y) = Cov(X, X) + Cov(X, Q)$$



Since $X$ and $Q$ are independent, $Cov(X, Q) = 0$. The covariance of $X$ with itself is its variance:

$$Cov(X, Y) = Var(X) + 0 = 1$$


* **Joint Distribution:**
The vector $(X, Y)$ follows a bivariate normal distribution:

$$(X, Y) \sim \mathcal{N}\left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} 1 & 1 \\ 1 & 5 \end{pmatrix} \right)$$



### **2. Information Geometry**

* **Mathematical Reason:** $X$ is $\sigma(Y)$-measurable if and only if $X$ can be written as a deterministic function of $Y$ (i.e., $X = g(Y)$). Because $Y = X + Q$, knowing $Y$ leaves us with an equation with two unknowns. Due to the presence of the independent random noise $Q$, there is no unique, deterministic mapping from $Y$ back to $X$.
* **Practical Terms:** If the engineer reads a high power spike of **$Y=5$**, they face an inverse problem. This spike could mean the motor is critically overheating ($X=5$) while cutting soft material ($Q=0$), or it could mean the motor is perfectly cool ($X=0$) but the CNC is cutting an extremely hard pocket of material ($Q=5$). The single measurement cannot separate the two independent sources.

### **3. Conditional Expectation (The Best Estimate)**

We use the standard property for jointly normal random variables to find $E[X \mid Y]$:


$$E[X \mid Y=y] = \mu_X + \frac{Cov(X, Y)}{Var(Y)} (y - \mu_Y)$$


Plugging in our derived values ($\mu_X = 0$, $\mu_Y = 0$, $Cov(X,Y) = 1$, $Var(Y) = 5$):


$$E[X \mid Y=y] = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$$


Therefore, the conditional expectation random variable is:


$$E[X \mid Y] = \frac{1}{5}Y$$

* **Signal-to-Total-Variance Ratio:**
The coefficient $\frac{1}{5}$ perfectly represents this ratio. The "Signal" variance is $Var(X) = 1$, and the "Total" variance is $Var(Y) = 5$. Thus, our best estimate essentially scales the observed reading by the fraction of the total variance that actually comes from the signal.

### **4. Verification of the Residual**

Let the estimator be $\hat{X} = E[X \mid Y] = \frac{1}{5}Y$. The residual (estimation error) is $Z = X - \hat{X}$.

* **Prove $E[Z] = 0$:**

$$Z = X - \frac{1}{5}Y = X - \frac{1}{5}(X + Q) = \frac{4}{5}X - \frac{1}{5}Q$$


$$E[Z] = \frac{4}{5}E[X] - \frac{1}{5}E[Q] = \frac{4}{5}(0) - \frac{1}{5}(0) = 0$$


* **Verify $Z$ is uncorrelated with $Y$ (Orthogonality Principle):**

$$Cov(Z, Y) = Cov(X - \frac{1}{5}Y, Y) = Cov(X, Y) - \frac{1}{5}Cov(Y, Y)$$


$$Cov(Z, Y) = Cov(X, Y) - \frac{1}{5}Var(Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0$$



Since the covariance is zero, the estimation error is perfectly uncorrelated with our observation.

### **5. Information Theory: Differential Entropy**

The differential entropy for a Gaussian distribution $\mathcal{N}(\mu, \sigma^2)$ in bits is $h = \frac{1}{2} \log_2(2\pi e \sigma^2)$.

* **Differential Entropy $h(X)$:**
With $\sigma_X^2 = 1$:

$$h(X) = \frac{1}{2} \log_2(2\pi e (1)) \approx 2.047 \text{ bits}$$


* **Conditional Differential Entropy $h(X \mid Y)$:**
First, compute the conditional variance $\sigma_{X|Y}^2$:

$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{Cov(X, Y)^2}{Var(Y)} = 1 - \frac{1^2}{5} = 1 - 0.2 = 0.8$$


$$h(X \mid Y) = \frac{1}{2} \log_2(2\pi e (0.8)) \approx 1.885 \text{ bits}$$



### **6. Mutual Information**

* **Calculate $I(X;Y)$:**

$$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2} \log_2(2\pi e) - \frac{1}{2} \log_2(2\pi e \cdot 0.8)$$


$$I(X; Y) = \frac{1}{2} \log_2\left(\frac{1}{0.8}\right) = \frac{1}{2} \log_2(1.25) \approx 0.161 \text{ bits}$$


* **Percentage of uncertainty removed:**
In continuous statistics, "uncertainty removed" often points to the reduction in variance ($R^2$):

$$\frac{Var(X) - Var(X|Y)}{Var(X)} = \frac{1 - 0.8}{1} = 20\%$$



* **If noise $Q$ variance was 100 instead of 4:**
Then $Var(Y) = 1 + 100 = 101$.
The conditional variance becomes $\sigma_{X|Y}^2 = 1 - \frac{1^2}{101} = \frac{100}{101} \approx 0.99$.
The mutual information drops to $\frac{1}{2} \log_2(1.01) \approx 0.007 \text{ bits}$.
Practically **0%** (specifically **~0.99%**) of the uncertainty would be removed, as the sensor reading would be overwhelmingly dominated by mechanical noise rather than temperature.

### **7. Diagnostic Interpretation**

The factory manager's intuition ("just use $Y$ to estimate $X$") is mathematically flawed in a conditional sense.

If we blindly estimate $\hat{X} = Y$, the conditional bias is massive:


$$E[X - Y \mid Y=y] = \frac{1}{5}y - y = -\frac{4}{5}y$$


This means for any given reading $y$, predicting the temperature is exactly $y$ will consistently overestimate the temperature by $80\%$ of that reading.

**Why does the optimal estimate "shrink" to the mean?** Because of the prior distributions, extreme temperatures (which have a low variance of 1) are very rare. Mechanical torque noise (which has a higher variance of 4) generates extreme values far more frequently. Therefore, if the sensor shows an incredibly high spike, probability dictates it is far more likely that the noise $Q$ spiked rather than the temperature $X$. The $E[X \mid Y]$ estimator mathematically optimally "shrinks" the estimate by $80\%$ towards the safest prior assumption (the mean, $0$) to protect against the high volatility of the mechanical noise.
