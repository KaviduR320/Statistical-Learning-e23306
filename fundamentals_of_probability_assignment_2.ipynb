#Q3) Transformer Failiures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.





### 1. Construct a Probability Space $(\Omega, \mathcal{F}, P)$

A probability space consists of the sample space, the $\sigma$-algebra, and the probability measure.

* **Sample Space ($\Omega$):** The set of all possible outcomes. Since the number of failures can be 0, 1, 2, or 3:

$$\Omega = \{0, 1, 2, 3\}$$


* **$\sigma$-algebra ($\mathcal{F}$):** For a finite sample space, we typically use the power set, which contains all possible subsets of $\Omega$:

$$\mathcal{F} = \mathcal{P}(\Omega)$$


* **Probability Measure ($P$):** Defined by the given values for each elementary outcome:

$$P(\{0\}) = 0.50, P(\{1\}) = 0.30, P(\{2\}) = 0.15, P(\{3\}) = 0.05$$



### 2. Define $X$ as a Random Variable

A random variable is a function $X: \Omega \to \mathbb{R}$. In this specific space where the outcomes are already numerical values representing the count of failures, $X$ is the **identity function**:


$$X(\omega) = \omega, \quad \text{for each } \omega \in \Omega$$

### 3. Probability Mass Function (PMF)

The PMF, $p_X(x)$, gives the probability that $X$ is exactly equal to some value $x$.

|$x$  | p_X(x)=P(X=x)|

| 0 | 0.50 |

| 1 | 0.30 |

| 2 | 0.15 |

| 3 | 0.05 |

### 4. Expectation $E[X]$

The expectation is the weighted average of all possible values:


$$E[X] = \sum x \cdot P(X=x)$$

$$E[X] = (0 \times 0.50) + (1 \times 0.30) + (2 \times 0.15) + (3 \times 0.05)$$

$$E[X] = 0 + 0.30 + 0.30 + 0.15 = \mathbf{0.75}$$

### 5. Variance $Var(X)$

First, find $E[X^2]$:


$$E[X^2] = (0^2 \times 0.50) + (1^2 \times 0.30) + (2^2 \times 0.15) + (3^2 \times 0.05)$$

$$E[X^2] = 0 + 0.30 + 0.60 + 0.45 = 1.35$$


Now, use the formula $Var(X) = E[X^2] - (E[X])^2$:


$$Var(X) = 1.35 - (0.75)^2 = 1.35 - 0.5625 = \mathbf{0.7875}$$

### 6. Probabilities $P(X \geq 1)$ and $P(X \geq 2)$

* **$P(X \geq 1)$:** This is the probability of at least one failure.

$$P(X \geq 1) = P(X=1) + P(X=2) + P(X=3) = 0.30 + 0.15 + 0.05 = \mathbf{0.50}$$



*(Alternatively: $1 - P(X=0) = 1 - 0.50 = 0.50$)*
* **$P(X \geq 2)$:**

$$P(X \geq 2) = P(X=2) + P(X=3) = 0.15 + 0.05 = \mathbf{0.20}$$



### 7. Practical Meaning

* **Expectation ($E[X] = 0.75$):** This represents the **long-term average** number of failures per day. If you monitored this district over many years, you would expect an average of 0.75 failures daily. It helps in resource planning and budgeting for repairs.
* **Variance ($Var(X) = 0.7875$):** This measures the **uncertainty or spread** of the failures. A higher variance would mean the number of daily failures fluctuates wildly (some days zero, some days many), making it harder to predict daily staffing needs compared to a low-variance scenario.

This question models the proportion of defective items $X$ using a Beta distribution with parameters $\alpha=2$ and $\beta=18$.
1.  Probability Space $(\Omega, \mathcal{F}, P)$:Sample Space ($\Omega$): The interval of all possible proportions, $\Omega = [0, 1]$.  $\sigma$-algebra ($\mathcal{F}$): The Borel $\sigma$-algebra on $[0, 1]$, denoted $\mathcal{B}([0, 1])$.  Probability Measure ($P$): Defined by the density function $f(x)$ such that for any event $A \in \mathcal{F}$, $P(A) = \int_A f(x) dx$.
2.  $X$ as a Random Variable:$X$ is the identity mapping $X(\omega) = \omega$ for $\omega \in [0, 1]$, which is a continuous random variable because its range is an interval and its distribution is defined by a density function.
3.  Probability Density Function (PDF):$$f(x) = \frac{1}{B(2,18)}x^{2-1}(1-x)^{18-1} = \frac{1}{B(2,18)}x(1-x)^{17} \quad \text{for } 0 < x < 1 \text{ [cite: 11, 19]}                                                                            $$
4. Verification:
A function is a valid PDF if it is non-negative and integrates to 1. Since $x$ and $(1-x)$ are positive on $(0, 1)$, $f(x) \geq 0$. By the definition of the Beta function:
  $$\int_0^1 x^{\alpha-1}(1-x)^{\beta-1} dx = B(\alpha, \beta)$$Thus, $\int_0^1 \frac{1}{B(2,18)}x(1-x)^{17} dx = \frac{B(2,18)}{B(2,18)} = 1$.
5.  Expectation $E[X]$:
For a Beta distribution, $E[X] = \frac{\alpha}{\alpha + \beta}$.
  $$E[X] = \frac{2}{2 + 18} = \frac{2}{20} = \mathbf{0.10}
$$
6. Variance $Var(X)$:$Var(X) = \frac{\alpha\beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}$.
  $$Var(X) = \frac{2 \times 18}{(20)^2(21)} = \frac{36}{400 \times 21} = \frac{36}{8400} = \mathbf{0.004286}
$$
7. $P(X \leq 0.10)$:$$P(X \leq 0.10) = \frac{1}{B(2,18)} \int_0^{0.10} x(1-x)^{17} dx
$$
8. $P(0.05 \leq X \leq 0.15)$:$$P(0.05 \leq X \leq 0.15) = \frac{1}{B(2,18)} \int_{0.05}^{0.15} x(1-x)^{17} dx
$$
9. Reasonableness of Model:
The Beta distribution is ideal because its support is strictly bounded between 0 and 1, matching the nature of a "proportion" or "percentage". It is also flexible enough to model skewed data, such as a process where defects are expected to be rare (low $\alpha$ relative to $\beta$).  
10. Interpretation:Expectation (0.10): The long-term average defective rate is $10\%$.  Variance: Measures the consistency of production quality; a low variance indicates a stable process where the daily defect rate stays close to $10\%$.

## Q6) Expected Project Completion Time Using PERT

This question focuses on the **Program Evaluation and Review Technique (PERT)** distribution, which is a version of the Beta distribution scaled to a specific interval $[a, b]$. In this case, the software integration task has a minimum time $a=8$ days, a maximum time $b=20$ days, and a most likely time $c=11$ days.

---

### 1. Compute the shape parameters $\alpha$ and $\beta$

The shape parameters are calculated using the formulas provided in the model:

*
**$\alpha$:** $1 + 4\left(\frac{c - a}{b - a}\right) = 1 + 4\left(\frac{11 - 8}{20 - 8}\right) = 1 + 4\left(\frac{3}{12}\right) = 1 + 1 = \mathbf{2}$.


*
**$\beta$:** $1 + 4\left(\frac{b - c}{b - a}\right) = 1 + 4\left(\frac{20 - 11}{20 - 8}\right) = 1 + 4\left(\frac{9}{12}\right) = 1 + 3 = \mathbf{4}$.



### 2. Support of $X$

The support of the random variable $X$ is the interval of all possible durations:
**Support: $[8, 20]$**.

### 3. Probability Density Function $f_X(x)$

Using the general formula for the PERT PDF:


$$f_X(x) = \frac{(x-8)^{2-1}(20-x)^{4-1}}{B(2, 4)(20-8)^{2+4-1}} = \mathbf{\frac{(x-8)(20-x)^3}{B(2, 4) \cdot 12^5}}$$

for $8 \le x \le 20$, where $B(2, 4)$ is the Beta function.

### 4. Probability Space $(\Omega, \mathcal{F}, P)$

*
**$\Omega$:** The interval $[8, 20]$ representing possible days.


*
**$\mathcal{F}$:** The Borel $\sigma$-algebra $\mathcal{B}([8, 20])$.


*
**$P$:** The probability measure defined by the density $f_X(x)$.



### 5. Define $X$ as a Random Variable

$X$ is defined as the identity function $X(\omega) = \omega$ for all $\omega \in \Omega$, mapping the outcome (number of days) to its real value.

### 6. Show that $E[X] = \frac{a+4c+b}{6}$

Using the specific parameters for this task:


$$E[X] = \frac{8 + 4(11) + 20}{6} = \frac{8 + 44 + 20}{6} = \frac{72}{6} = \mathbf{12 \text{ days}}$$

.

### 7. Compute the Variance $Var(X)$

Using the approximation provided:


$$Var(X) = \frac{(E[X] - a)(b - E[X])}{7} = \frac{(12 - 8)(20 - 12)}{7} = \frac{4 \times 8}{7} = \frac{32}{7} \approx \mathbf{4.57}$$

.

### 8. Approximate Standard Deviation $\sigma$

$$\sigma = \sqrt{Var(X)} = \sqrt{4.57} \approx \mathbf{2.14 \text{ days}}$$

.

### 9. PERT vs. Triangular Distribution

The PERT distribution is often considered more appropriate because it is **smoother** and places more weight on the "most likely" value $(c)$ compared to the Triangular distribution. In a Triangular model, the mean is $\frac{8+20+11}{3} = 13$, whereas PERT gives $12$. PERT reduces the influence of extreme values (the "tails"), which often leads to more realistic project estimates.

### 10. Practical Interpretation

*
**$a, b, c$:** Represent the optimistic, pessimistic, and most likely time estimates provided by experts.


*
**$E[X]$:** The weighted average used for scheduling; it accounts for the fact that extreme delays $(b)$ are possible but less likely than the mode.


*
**$\sigma$:** Represents the risk or uncertainty; a larger $\sigma$ indicates a more unpredictable task that might require a larger time buffer in the project plan.

## Q7) If the PERT Expectation is to Match Our Intuition

This question explores a specific scenario where the statistical expectation ($E[X]$) of the PERT model aligns exactly with the most likely time ($c$) provided by expert judgment. Given the parameters $a = 5$ (minimum time) and $c = 11$ (most likely time):

### 1. Determine the value of the maximum time $b$

To find $b$, we set the PERT expectation formula equal to the most likely time $c$:

$$E[X] = \frac{a + 4c + b}{6} = c$$

Plugging in the known values:


$$11 = \frac{5 + 4(11) + b}{6}$$

$$11 = \frac{5 + 44 + b}{6}$$

$$66 = 49 + b$$

$$b = 66 - 49$$

$$\mathbf{b = 17 \text{ days}}$$

For the expected time to match the most likely time, the maximum possible duration must be **17 days**.

---

### 2. What happens to the variance?

The variance in a PERT distribution is a measure of the spread or uncertainty of the task duration.

*
**Symmetry:** When $E[X] = c$, the distribution becomes perfectly symmetric around the mode. In this case, the distance from the minimum to the mode ($11 - 5 = 6$) is equal to the distance from the mode to the maximum ($17 - 11 = 6$).


*
**Calculation:** Using the approximation formula $Var(X) = \frac{(E[X] - a)(b - E[X])}{7}$:



$$Var(X) = \frac{(11 - 5)(17 - 11)}{7}$$


$$Var(X) = \frac{6 \times 6}{7} = \frac{36}{7} \approx \mathbf{5.14}$$



**Practical Meaning:** In this balanced scenario, the variance reaches a "stable" point relative to the range. Because the distribution is no longer skewed toward a very long "worst-case" scenario or a very short "best-case" scenario, the uncertainty is distributed evenly on both sides of the expected completion date.
